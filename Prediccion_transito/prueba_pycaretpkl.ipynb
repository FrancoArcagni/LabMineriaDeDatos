```python
from pycaret.datasets import get_data
```

```python
data1 = get_data(dataset = 'traffic')
```

Queremos predecir cuanto trafico vamos a tener dependiendo de ciertas variables del clima. La variable target es traffic y vemos que es un modelo de regresion porque se trata de predecir una variable continua (numerica).
Como sabemos que es una regresion importamos la siguiente libreria

```python
from pycaret.regression import *
```

```python
dataset = setup(data = data1, target = 'traffic_volume', session_id = 123)
#en los entornos no productivos o de desarrollo se recomieda siempre usar el session_id con cualquier numero.
#Esta es la variable que me inicializa el generador de numeros aleatorios. Y si le doy un numero el generador siempre va a tener
#el mismo comportamiento. Por lo tanto, si yo vuelvo a correr el setup, el resultado va a ser el mismo.
#En entornos productivos, se recomienda no usar el session_id, para que el generador de numeros aleatorios tenga un comportamiento diferente
#en cada ejecucion. Esto es para que el modelo no se vea afectado por el mismo comportamiento de los numeros aleatorios.
```

```python
best = compare_models()
#El compare_models me va a devolver el mejor modelo.
```

```python
tuned_best = tune_model(estimator = best)
#El tune_model me va a devolver el mejor modelo ajustado. Es para optimizarle los hiperparametros al mejor modelo
```

```python
final_model = finalize_model(estimator = tuned_best)
#El finalize_model me va a devolver el mejor modelo ajustado y entrenado. Es para que yo pueda usar el modelo para hacer predicciones
```

```python
save_model(model = final_model, model_name = 'prueba_pycaret.pkl')
#Guardamos el modelo listo para pasar a produccion
```
